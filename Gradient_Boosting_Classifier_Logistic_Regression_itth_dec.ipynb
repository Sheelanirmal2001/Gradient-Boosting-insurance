```python
from sklearn.model_selection import train_test_split
xtrain_full,xtest,ytrain_full,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# within training data split into val data
# early stopping rounds
xtrain,xval,ytrain,yval=train_test_split(xtrain_full,ytrain_full,random_state=0)
```

XGBoost Classifier

```python
from xgboost import XGBClassifier
```

```python
xgbc=XGBClassifier(n_estimators=500,learning_rate=0.05,use_label_encoder=False,eval_metric='logloss',early_stopping_rounds=5,n_jobs=-1)
```

```python
xgbc.fit(xtrain,ytrain,eval_set=[(xval,yval)])
```

```python
from sklearn.metrics import accuracy_score
```

```python
ypred=xgbc.predict(xtest)
```

```python
import numpy as np
print(np.round(accuracy_score(ypred,ytest),2))
```
